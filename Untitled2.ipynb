```python
# change campus names
def change_school_name(df):
    
    # school name dictionary
    school_dict = {
        '/': '',
        'Bilingual': '',
        'Non-LEP DL': '',
        'ESOL': '',
        'Regular': '',
        'Self Contained': '',
        'High': '',
        'Elem School': '',
        'Elementary': '',
        'Middle': '',
        'School': '',
        'HS': '',
        'ES': '',
        'MS': '',
        'Tr.': 'Trail',
        ' - ': ' ',
        'Daep': 'DAEP',
        'Juvenile Ctr': '',
        'Wm.': '',
        'Wm S': '',
        'Xenia': '',
        'Liveoak': 'Live Oak',
        'Linda Herrington': 'Herrington',
        'Elsa ': '',
        'Linda': 'Herrington',
        'C. D. Fulkes': 'C.D. Fulkes',
        'RROC': 'Round Rock Opportunity Center',
        'RRISD ': '',
        'Joe Lee ': '',
        'James Walsh': 'Walsh',
        'Neysa Callison': 'Callison',
        'Kathy Caraway': 'Caraway',
        'Noel Grisham': 'Grisham',
        'Claude Berkman': 'Berkman',
        'Patsy Sommer': 'Sommer',
        '\t\r': '',
        '  ': ' ',
        'Lott': 'LOTT',
        'Elem: DAEP': 'DAEP',
        'Elem: ': '',
        'Mt.': 'Mountain',
    }
    
    # change keys(k) to values(v)
    for k, v in school_dict.items():
        df['School'] = df['School'].map(lambda dis_cell: dis_cell.replace(k, v).strip())
    
    return None
```
